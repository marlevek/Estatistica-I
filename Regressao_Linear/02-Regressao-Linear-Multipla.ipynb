```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
base = pd.read_csv('mt_cars.csv')
base.head()
```

```python
base = base.drop(['Unnamed: 0'], axis=1)
base.head()
```

```python
# Criação de variáveis X e y e modelo para prever mpg: cyl - disp - hp
X = base.iloc[:, 1:4].values
y = base.iloc[:, 0].values
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
modelo.score(X, y)
```

```python
# Criação do modelo ajustado com mais atributos (Regressão Linear Múltipla)
modelo_ajustado = sm.ols(formula = 'mpg ~ cyl + disp + hp', data = base)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
# Previsão de um novo registro
novo = np.array([4, 200, 100])
novo = novo.reshape(1, -1)
modelo.predict(novo)
```
